```python
import numpy as np
import matplotlib.pyplot as plt

# Setting random seed so we get the same results every time
np.random.seed(0)

# ---- Uniform Distribution ----

population = np.random.uniform(low=0, high=100, size=10000)
sample_means = []
for i in range(1000):
    sample = np.random.choice(population, size=30, replace=False)
    sample_means.append(np.mean(sample))

# Plot
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.hist(population, bins=50, color='lightblue', edgecolor='black')
plt.title('Population Distribution (Uniform)')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.subplot(2, 1, 2)
plt.hist(sample_means, bins=30, color='lightgreen', edgecolor='black')
plt.title('Sampling Distribution of the Sample Mean (Uniform)')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

# ---- Exponential Distribution ----

population2 = np.random.exponential(scale=5, size=10000)
sample_means2 = []
for i in range(1000):
    sample = np.random.choice(population2, size=30, replace=False)
    sample_means2.append(np.mean(sample))

# Plot
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.hist(population2, bins=50, color='lightblue', edgecolor='black')
plt.title('Population Distribution (Exponential)')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.subplot(2, 1, 2)
plt.hist(sample_means2, bins=30, color='lightgreen', edgecolor='black')
plt.title('Sampling Distribution of the Sample Mean (Exponential)')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

# ---- Binomial Distribution ----
# Creating a population using binomial distribution

population3 = np.random.binomial(n=10, p=0.5, size=10000)
# Taking 1000 samples and store the mean of each one
sample_means3 = []
for i in range(1000):
    sample = np.random.choice(population3, size=30, replace=False)
    sample_means3.append(np.mean(sample))

# Plot
plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.hist(population3, bins=50, color='lightblue', edgecolor='black')
plt.title('Population Distribution (Binomial)')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')

plt.subplot(2, 1, 2)
plt.hist(sample_means3, bins=30, color='lightgreen', edgecolor='black')
plt.title('Sampling Distribution of the Sample Mean (Binomial)')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
