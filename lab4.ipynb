```python
!pip install faker
import pandas as pd
import numpy as np
import random
from faker import Faker

fake = Faker()
np.random.seed(42)
random.seed(42)
n = 100
dates = []
for i in range(n):
    if random.random() > 0.1:
        dates.append(fake.date_this_decade())
    else:
        dates.append(None)  # 10% chance of missing date

customer_ids = []
for i in range(n):
    if random.random() > 0.05:
        customer_ids.append(random.randint(1000, 9999))
    else:
        customer_ids.append(None)  # 5% chance of missing customer ID

products = []
for i in range(n):
    products.append(random.choice(['Laptop', 'Phone', 'Tablet', 'Monitor', 'Keyboard', 'Headphones', 'Mouse']))

categories = []
for i in range(n):
    categories.append(random.choice(['Electronics', 'Furniture', 'Toys', 'Clothing', 'Sports']))

sales = []
for i in range(n):
    if random.random() > 0.1:
        sales.append(round(random.uniform(50, 1500), 2))
    else:
        sales.append(None)  # 10% chance of missing sales value

df = pd.DataFrame({
    'Date': dates,
    'CustomerID': customer_ids,
    'Product': products,
    'Category': categories,
    'Sales': sales
})

for i in range(len(df)):
    if random.random() < 0.1:
        df.loc[i, 'Product'] = None

duplicates = df.sample(frac=0.05, random_state=42)
df = pd.concat([df, duplicates], ignore_index=True)

# Save to CSV
df.to_csv('sales_data.csv', index=False)

print(df.head())
print(df.info())
```
